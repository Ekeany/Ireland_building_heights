```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from BorutaShap import BorutaShap

#mean_squared_error(y_actual, y_predicted, squared=False)

file = 'C:/Users/egnke/PythonCode/Met_Eireann_git/Ireland_building_heights/Modelling/DataSetCreation/building_height_sample.csv'
data = pd.read_csv(file)
print(data.shape)
data.head()
```

```python
data = data.sample(40000)
y = data['Height']
X = data.drop(['X','Y','tile','Height'], axis=1)


# if classification is False it is a Regression problem
Feature_Selector = BorutaShap(importance_measure='shap',
                              classification=False)


Feature_Selector.fit(X=X, y=y, n_trials=50,
                     train_or_test = 'test', normalize=True,
                     verbose=True)


subset = Feature_Selector.Subset()
subset.to_csv('subset.csv',index=False)
subset.head()
```
