```python
def get_polygon(string_):
    
    between_brackets = string_[string_.find("(")+3:string_.find(")))")]
    between_brackets = between_brackets.replace(')','').replace('(','')
    
    long_lat = []
    for str_coord in between_brackets.split(','):

        long_lat.append(str_coord.lstrip().split(' '))
    
    long_lat = [list(map(float, sublist)) for sublist in long_lat]
    
    poly = []
    for elem in long_lat:
        x, y = elem
        poly.append((x,y))

    
    return Polygon(poly)


def get_interesction(poly1, poly2):
    # dividing by the area for completely interected polygons
    try:
        return poly2.intersection(poly1).area
    
    except:
        return -1


def normalize_the_intersection(df, column):
    maximum =  df[column].max()
    df[column] = df[column]/maximum
    return df
    
    

def plot_polygons(poly1, poly2):
    
    x,y = poly1.exterior.xy
    x1,y1 = poly2.exterior.xy
    plt.plot(x,y)
    plt.plot(x1,y1)
    

def flatten_list(array):
    return [item for sublist in array for item in sublist]

    
#dublin_poly = Polygon([(-6.6663, 53.5769),(-6.4968, 53.1487),(-5.8961, 53.2305),(-6.0599, 53.6596)])
#all_ireland_poly = Polygon([(-5.141389,55.369722),(-10.040833,55.478611),(-11.381944,51.041389),(-6.031111,51.522222)])

#unique_tiles['intersection_area'] = unique_tiles.footprint.apply(lambda x: get_interesction(get_polygon(x), all_ireland_poly))
#unique_tiles = normalize_the_intersection(unique_tiles, 'intersection_area')
#unique_tiles['intersection_area'].hist(bins=35)
#unique_tiles.footprint.apply(lambda x: plot_polygons(get_polygon(x), all_ireland_poly))   
    
    
cols = ['footprint','tileid']

# left to right similar to the map of ireland
tiles = [['29UNB','29UPB'],
         ['29UMA','29UNA','29UPA','30UUF'],
         ['29UMV','29UNV','29UPV'],
         ['29UMU','29UNU','29UPU'],
         ['29UMT','29UNT','29UPT']]


products_df['area'] = products_df.footprint.apply(lambda x: get_polygon(x).area)
products_df['centroid'] = products_df.footprint.apply(lambda x: get_polygon(x).centroid.coords)

tiles_of_interest = products_df[products_df['tileid'].isin(flatten_list(tiles))]
tiles_of_interest = tiles_of_interest.groupby('tileid', group_keys=False).apply(lambda x: x.loc[x['area'].idxmax()])
tiles_of_interest
```

```python
def format_coords(string_):
    
    between_brackets = string_[string_.find("(")+3:string_.find(")))")]
    between_brackets = between_brackets.replace(')','').replace('(','')
    
    long_lat = []
    for str_coord in between_brackets.split(','):

        long_lat.append(str_coord.lstrip().split(' '))
    
    long_lat = [list(map(float, sublist)) for sublist in long_lat]
    
    poly = []
    for elem in long_lat:
        x, y = elem
        poly.append([x,y])

    
    return poly


def create_geo_json_feature(coords, name):
    coords = format_coords(coords)
    
    geo = {
        "type": "Feature",
        "properties": {"tile_name": name},
        "geometry": {
            "type": "Polygon",
            "coordinates": [coords]
        }
    }

    return geo


def create_all_tiles_map(df, m, filename='all_tiles.html'):
    
    colors = ['red', 'blue', 'green', 
             'purple', 'orange', 'darkred', 
             'lightred', 'beige', 'darkblue', 
             'darkgreen', 'cadetblue', 'darkpurple', 'white', 
             'pink', 'lightblue', 'lightgreen', 'gray', 
             'black', 'lightgray']
    
    tile_names = df.tileid.values
    footprints = df.footprint.values
    for footprint, tile in zip(footprints, tile_names):
        
        geo = create_geo_json_feature(footprint, tile)
        color = random.choice(colors)
        
        g = folium.GeoJson(
            geo,
            style_function=lambda feature: {
                    'fillColor':color,
                    'color':color,
                    'fillOpacity':0.5
                },   
        ).add_to(m)

        folium.GeoJsonTooltip(fields=["tile_name"]).add_to(g)
        
    m.save(filename)  

    
m = folium.Map(location=[37, 0],
               zoom_start=2.5,
               tiles='https://server.arcgisonline.com/arcgis/rest/services/Canvas/World_Dark_Gray_Base/MapServer/tile/{z}/{y}/{x}',
               attr='My Data Attribution')
    
create_all_tiles_map(tiles_of_interest, m)
```

```python
def create_small_bounding_box(top_left_x, top_left_y):
    
    top_right_x = top_left_x + 0.05
    top_right_y = top_left_y
    
    bottom_left_x = top_left_x
    bottom_left_y = top_left_y - 0.05
    
    bottom_right_x = top_left_x + 0.05
    bottom_right_y = top_left_y - 0.05
    
    #return f'{top_left_x}/{top_left_y},{bottom_right_x}/{bottom_right_y},{top_right_x}/{top_right_y},{bottom_right_x}/{bottom_right_y}'
    return f'{top_left_x}/{top_left_y},{bottom_left_x}/{bottom_left_y},{bottom_right_x}/{bottom_right_y},{top_right_x}/{top_right_y},{top_left_x}/{top_left_y}'


def format_coord_center_points(string_):
    

    long_lat = []
    for str_coord in string_.split(','):

        long_lat.append(str_coord.lstrip().split('/'))
    
    long_lat = [list(map(float, sublist)) for sublist in long_lat]
    
    poly = []
    for elem in long_lat:
        x, y = elem
        poly.append([x,y])

    
    return poly    


def create_geo_json_feature(coords, name):
    
    coords = format_coord_center_points(coords)
    
    geo = {
        "type": "Feature",
        "properties": {"tile_name": name},
        "geometry": {
            "type": "Polygon",
            "coordinates": [coords]
        }
    }

    return geo


def create_tile_center_map(df, m, filename='tiles_center_point.html'):
    
    colors = ['red', 'blue', 'green', 
             'purple', 'orange', 'darkred', 
             'lightred', 'beige', 'darkblue', 
             'darkgreen', 'cadetblue', 'darkpurple', 'white', 
             'pink', 'lightblue', 'lightgreen', 'gray', 
             'black', 'lightgray']
    
    tile_names = df.tileid.values
    center_points = df.center_points.values
    for center_point, tile in zip(center_points, tile_names):
        
        geo = create_geo_json_feature(center_point, tile)
        color = random.choice(colors)
        
        g = folium.GeoJson(
            geo,
            style_function=lambda feature: {
                    'fillColor':color,
                    'color':color,
                    'fillOpacity':0.5
                },   
        ).add_to(m)

        folium.GeoJsonTooltip(fields=["tile_name"]).add_to(g)
        
    m.save(filename)  

    
m = folium.Map(location=[37, 0],
               zoom_start=2.5,
               tiles='https://server.arcgisonline.com/arcgis/rest/services/Canvas/World_Dark_Gray_Base/MapServer/tile/{z}/{y}/{x}',
               attr='My Data Attribution')

    
tiles_of_interest['center_points'] = tiles_of_interest['centroid'].apply(lambda shap_coord: create_small_bounding_box(shap_coord[0][0],shap_coord[0][1]))
create_tile_center_map(tiles_of_interest, m)
```
